```python
import math

class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __str__(self):
        return f"<{self.x}, {self.y}, {self.z}>"

    def __add__(self, otro):
        return Vector3D(self.x + otro.x, self.y + otro.y, self.z + otro.z)

    def __sub__(self, otro):
        return Vector3D(self.x - otro.x, self.y - otro.y, self.z - otro.z)

    def __mul__(self, k):
        return Vector3D(self.x * k, self.y * k, self.z * k)

    def __rmul__(self, k):
        return self.__mul__(k)

    def __eq__(self, otro):
        return self.x == otro.x and self.y == otro.y and self.z == otro.z

    def __neg__(self):
        return Vector3D(-self.x, -self.y, -self.z)

    def producto_punto(self, otro):
        return self.x * otro.x + self.y * otro.y + self.z * otro.z

    def es_ortogonal(self, otro):
        return self.producto_punto(otro) == 0

    def magnitud(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

    def angulo(self, otro):
        cos_theta = self.producto_punto(otro) / (self.magnitud() * otro.magnitud())
        return math.acos(cos_theta)


v1 = Vector3D(1, 0, 1)
v2 = Vector3D(0, 1, 0)
v3 = Vector3D(1, 0, 1)

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Suma: {v1 + v2}")
print(f"Resta: {v1 - v2}")
print(f"Multiplicación por escalar: {2 * v1}")
print(f"Inverso aditivo de v2: {-v2}")
print(f"¿v1 y v3 son iguales?: {v1 == v3}")
print(f"Producto punto: {v1.producto_punto(v2)}")
print(f"¿Son ortogonales?: {v1.es_ortogonal(v2)}")
print(f"Magnitud de v1: {v1.magnitud():.3f}")
print(f"Ángulo entre v1 y v2: {v1.angulo(v2):.3f} rad")
```
